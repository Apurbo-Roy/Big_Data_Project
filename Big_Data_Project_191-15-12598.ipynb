```python
from random import randint
def randomnumber(n):
    first = 10**(n-1)
    end = (10**n)-1
    return randint(first,end)


table3 = spark.read.csv("Downloads/Data Set - Sheet1.csv", header=True, inferSchema= True)

table3 = table3.withColumn("User ID", f.col("Emp ID"))\
        .withColumn("Tag", f.concat(f.lit(randomnumber(8))))\
        .withColumn("Activated", lit("1"))\
        .withColumn("Created_date", lit("05/29/2021 11:01:31"))\
        .withColumn("Updated_date", lit("05/29/2021 11:01:31"))\
        .withColumn("Valid_until", lit("05/29/2021 11:01:31'"))\
        .withColumn("Last_used", lit("05/29/2021 11:01:31"))\
        .withColumn("Employee_code", f.col("Emp ID"))\
        .withColumn("Reference_number", f.col("Employee_code"))

table3 = table3.withColumn("Unique_user_id", f.col('User ID') +1512598)
        
table3 = table3.select("User ID","Tag","Activated","Created_date","Updated_date","Valid_until","Last_used","Reference_number","Employee_code","Unique_user_id")\
        .drop("Employee Name","RFID Card Number","SL No.", "Contact")

table3 = table3.toPandas().to_csv('OneDrive/Desktop/users_rfidinfo.csv', header=True, index=False)
```

```python
table3_schema = StructType()

table3_schema = table3_schema.add("user_id",StringType(),True)\
              .add("tag",StringType(),True)\
              .add("activated",IntegerType(),True)\
              .add("created_date",StringType(),True)\
              .add("updated_date",StringType(),True)\
              .add("valid_until",IntegerType(),True)\
              .add("last_used",StringType(),True)\
              .add("reference_number",StringType(),True)\
              .add("employee_code",StringType(),True)\
              .add("unique_user_id",StringType(),True)

users_rfidinfo = spark.read.load("OneDrive/Desktop/users_rfidinfo.csv", format ="csv", sep = ",", inferSchema= True, header = True )
```
